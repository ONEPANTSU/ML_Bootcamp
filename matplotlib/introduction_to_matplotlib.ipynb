## Making figures with NumPy arrays
___
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a Bar plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Price ($)");
```

```python
fig, ax =plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

### 1. Subplots option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

### 1. Subplots option 2

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

## Plotting from Pandas DataFrames

```python
# Make a DataFrame

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = pd.Series(map(int, car_sales["Price"].str[:-2].replace('[\$\,\.]', '', regex=True)))
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Let's plot the Total Sales

car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
car_sales.plot.bar()
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="bar");
```

```python
car_sales["Odometer (KM)"].plot(kind="hist", bins=10)
```

## Heart Disease

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True, bins=50);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
# Pyplot method
over_50.plot(figsize=(10, 6),
             kind="scatter", 
             x='age', 
             y="chol", 
             c="target");
```

```python
# OO method with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)
# ax.set_xlim([45, 100])
```

```python
# OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol Level")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(), 
           linestyle="--", 
           color="red")
```

## Subplot of chol, age, thalach

```python
fig, ax = plt.subplots(nrows=2, 
                       ncols=1, 
                       figsize=(10, 10),
                       sharex=True)
# Axes 0
scatter = ax[0].scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"])

ax[0].set(title="Heart Disease and Cholesterol Levels",
          ylabel="Cholesterol Level")

ax[0].legend(*scatter.legend_elements(), title="Target")

ax[0].axhline(over_50["chol"].mean(), 
           linestyle="--", 
           color="red")

# Axes 1
scatter = ax[1].scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"])

ax[1].set(title="Heart Disease and Max Heart Rate",
          xlabel="Age",
          ylabel="Thalach Level")

ax[1].legend(*scatter.legend_elements(), title="Target")

ax[1].axhline(over_50["thalach"].mean(), 
           linestyle="--", 
           color="red");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

## Customizing Matplotlib plots and gettinf stylish

```python
plt.style.available
```

```python
plt.style.use("default")
ax = car_sales.Price.plot()
ax.set(title="Price", xlabel="ID", ylabel="$")
ax.legend().set_visible(True);
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig, ax = plt.subplots(nrows=2, 
                       ncols=1, 
                       figsize=(10, 10),
                       sharex=True)
# Axes 0
scatter = ax[0].scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"],
                        cmap="Accent") # https://matplotlib.org/stable/tutorials/colors/colormaps.html

ax[0].set(title="Heart Disease and Cholesterol Levels",
          ylabel="Cholesterol Level")
ax[0].set_xlim([50, 80])
ax[0].legend(*scatter.legend_elements(), title="Target")

ax[0].axhline(over_50["chol"].mean(), 
           linestyle="--", 
           color="red")

# Axes 1
scatter = ax[1].scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"],
                        cmap="Accent") # https://matplotlib.org/stable/tutorials/colors/colormaps.html

ax[1].set(title="Heart Disease and Max Heart Rate",
          xlabel="Age",
          ylabel="Thalach Level")
ax[1].set_ylim([60, 200])
ax[1].legend(*scatter.legend_elements(), title="Target")

ax[1].axhline(over_50["thalach"].mean(), 
           linestyle="--", 
           color="red");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
